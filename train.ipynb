```python
import os
import shutil
import yaml

# -----------------------------
# 1. Load YAML file
# -----------------------------
yaml_path = "data.yaml"

with open(yaml_path) as f:
    data = yaml.safe_load(f)

class_names = data["names"]

# -----------------------------
# 2. Define classes we want
# -----------------------------

human_classes = [
    'Pedestrians', 'Persona',
    'Pessoa', 'people', 'person', 'persons'
]

animal_classes = [
    'bird', 'cat','cow', 'dog',
    'fox', 'goat', 'horse', 'racoon', 'sheep'
]

# -----------------------------
# 3. Convert class names → IDs
# -----------------------------

human_ids = [class_names.index(c) for c in human_classes if c in class_names]
animal_ids = [class_names.index(c) for c in animal_classes if c in class_names]

print("Human class IDs:", human_ids)
print("Animal class IDs:", animal_ids)

# -----------------------------
# 4. Dataset splits
# -----------------------------

splits = ["train", "valid", "test"]

# -----------------------------
# 5. Process dataset
# -----------------------------

for split in splits:

    image_dir = f"{split}/images"
    label_dir = f"{split}/labels"

    new_image_dir = f"subset_dataset/{split}/images"
    new_label_dir = f"subset_dataset/{split}/labels"

    os.makedirs(new_image_dir, exist_ok=True)
    os.makedirs(new_label_dir, exist_ok=True)

    for label_file in os.listdir(label_dir):

        label_path = os.path.join(label_dir, label_file)

        with open(label_path) as f:
            lines = f.readlines()

        new_lines = []

        for line in lines:

            parts = line.strip().split()
            class_id = int(parts[0])

            # human → 0
            if class_id in human_ids:
                parts[0] = "0"
                new_lines.append(" ".join(parts))

            # animal → 1
            elif class_id in animal_ids:
                parts[0] = "1"
                new_lines.append(" ".join(parts))

        # if image has valid objects
        if new_lines:

            image_name = label_file.replace(".txt", ".jpg")
            src_img = os.path.join(image_dir, image_name)

            if not os.path.exists(src_img):
                continue

            shutil.copy(src_img, os.path.join(new_image_dir, image_name))

            with open(os.path.join(new_label_dir, label_file), "w") as f:
                f.write("\n".join(new_lines))

print("Subset dataset created successfully!")
```

```python
import os
import cv2
import random

# YOLO dataset paths
train_image_dir = "dataset/train/images"
train_label_dir = "dataset/train/labels"

val_image_dir = "dataset/valid/images"
val_label_dir = "dataset/valid/labels"

# Output dataset
output_dir = "dataset/classification_dataset"

train_human_dir = os.path.join(output_dir, "train/human")
train_animal_dir = os.path.join(output_dir, "train/animal")

val_human_dir = os.path.join(output_dir, "val/human")
val_animal_dir = os.path.join(output_dir, "val/animal")

os.makedirs(train_human_dir, exist_ok=True)
os.makedirs(train_animal_dir, exist_ok=True)
os.makedirs(val_human_dir, exist_ok=True)
os.makedirs(val_animal_dir, exist_ok=True)


def process_dataset(image_dir, label_dir, human_dir, animal_dir,
                    max_human=None, max_animal=None, image_subset=None):

    human_count = 0
    animal_count = 0

    images = os.listdir(image_dir)

    if image_subset is not None:
        images = image_subset

    random.shuffle(images)

    for img_name in images:

        if max_human and max_animal:
            if human_count >= max_human and animal_count >= max_animal:
                break

        img_path = os.path.join(image_dir, img_name)

        name = os.path.splitext(img_name)[0]
        label_path = os.path.join(label_dir, name + ".txt")

        if not os.path.exists(label_path):
            continue

        img = cv2.imread(img_path)
        if img is None:
            continue

        h, w, _ = img.shape

        with open(label_path) as f:
            lines = f.readlines()

        for line in lines:

            parts = line.strip().split()

            if len(parts) < 5:
                continue

            cls, x, y, bw, bh = map(float, parts[:5])

            x1 = int((x - bw/2) * w)
            y1 = int((y - bh/2) * h)
            x2 = int((x + bw/2) * w)
            y2 = int((y + bh/2) * h)

            x1 = max(0, x1)
            y1 = max(0, y1)
            x2 = min(w, x2)
            y2 = min(h, y2)

            crop = img[y1:y2, x1:x2]

            if crop.size == 0:
                continue

            # Ignore very small crops
            if crop.shape[0] < 40 or crop.shape[1] < 40:
                continue

            if int(cls) == 0 and (max_human is None or human_count < max_human):

                save_path = os.path.join(human_dir, f"human_{human_count}.jpg")
                cv2.imwrite(save_path, crop)
                human_count += 1

            elif int(cls) == 1 and (max_animal is None or animal_count < max_animal):

                save_path = os.path.join(animal_dir, f"animal_{animal_count}.jpg")
                cv2.imwrite(save_path, crop)
                animal_count += 1

    return human_count, animal_count


# ======================
# Create TRAIN dataset
# ======================

train_h, train_a = process_dataset(
    train_image_dir,
    train_label_dir,
    train_human_dir,
    train_animal_dir,
    max_human=2000,
    max_animal=2000
)

# ======================
# Create VAL dataset
# ======================

val_images = os.listdir(val_image_dir)
random.shuffle(val_images)

VAL_LIMIT = 200

val_h, val_a = process_dataset(
    val_image_dir,
    val_label_dir,
    val_human_dir,
    val_animal_dir,
    max_human=VAL_LIMIT,
    max_animal=VAL_LIMIT,
    image_subset=val_images
)

print("Train Humans:", train_h)
print("Train Animals:", train_a)

print("Val Humans:", val_h)
print("Val Animals:", val_a)

print("Classification dataset created successfully.")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
import wandb
```

```python
wandb.init(project="animal-human-classification")
```

```python
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485,0.456,0.406],
        std=[0.229,0.224,0.225]
    )
])
```

```python
train_dataset = datasets.ImageFolder("dataset/classification_dataset/train", transform=transform)
val_dataset = datasets.ImageFolder("dataset/classification_dataset/val", transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
model = models.resnet50(pretrained=True)

num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 2)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("Using device:", device)
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(
    model.parameters(),
    lr=0.0001
)
```

```python
epochs = 15

best_acc = 0
best_loss = float("inf")

for epoch in range(epochs):

    # -----------------
    # TRAINING
    # -----------------
    model.train()
    running_loss = 0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    train_loss = running_loss / len(train_loader)
    train_acc = correct / total

    # -----------------
    # VALIDATION
    # -----------------
    model.eval()
    val_loss = 0
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()

    val_loss /= len(val_loader)
    val_acc = val_correct / val_total

    print(
        f"Epoch {epoch+1}/{epochs} | "
        f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
        f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}"
    )

    # Log to wandb
    wandb.log({
        "train_loss": train_loss,
        "train_accuracy": train_acc,
        "val_loss": val_loss,
        "val_accuracy": val_acc
    })

    # Save best model based on validation
    if val_acc >= best_acc and val_loss <= best_loss:
        best_acc = val_acc
        best_loss = val_loss
        torch.save(model.state_dict(), "models/best_classification_model.pth")
        print("✅ Best model saved based on validation")
```
